```python
#Gender Pie chart scaffolding
male_sum = df_admit['Male'].sum()
female_sum = df_admit['Female'].sum()

trace = go.Pie(labels = ['Male','Female'], values = [male_sum,female_sum])
data = [trace]
layout = go.Layout(title="UC Firsty Years by Gender (2020)")
fig = go.Figure(data = data,layout = layout)
fig.show()
plotly.io.write_html(fig,"UC_GenderPie_2020.html", full_html=False)
```

```python
# Pie chart scaffolding
county = df["County/State/ Territory"].value_counts().keys().tolist()
countys_count = df["County/State/ Territory"].value_counts().values.tolist()
biggest_county_count = []
smallest_county_count = []

for i in countys_count:
    if i > 50:
        biggest_county_count.append(i)
    else:
        smallest_county_count.append(i)
        
print(smallest_county_count)    
trace = go.Pie(labels=county, 
                values=biggest_county_count, 
                marker=dict(colors=['red']), 
                hoverinfo="value"
              )
data = [trace]
layout = go.Layout(title="UC Firsty Years by County (2020)")
fig = go.Figure(data = data,layout = layout)
fig.show()
```

```python
plotly.io.write_html(fig,"UC_Pie_2020.html", full_html=False)
```
